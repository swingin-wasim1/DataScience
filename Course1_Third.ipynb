```python
def min_max(row):
    data = row[["POPESTIMATE2010",
               "POPESTIMATE2011",
               "POPESTIMATE2012",
               "POPESTIMATE2013",
               "POPESTIMATE2014",
               "POPESTIMATE2015"]]
    return pd.Series({'min':np.min(data),'max':np.max(data)})
df.apply(min_max,axis=1)
```

```python
rows =["POPESTIMATE2010",
               "POPESTIMATE2011",
               "POPESTIMATE2012",
               "POPESTIMATE2013",
               "POPESTIMATE2014",
               "POPESTIMATE2015"]
df.apply(lambda x:np.max(x[rows]),axis=1)
```

```python
df = pd.read_csv('census.csv')
for state in df['STNAME'].unique():
    avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
    print(state+' pop is '+str(avg))
    
                              
```

```python
for group,frame in df.groupby('STNAME'):
    avg =  np.average(frame['CENSUS2010POP'])
    print(group +" - "+str(avg))
```

```python
energy =pd.read_excel('Energy Indicators.xls',skiprows=17,skip_footer=38,index_col=2)
energy.replace('...', np.nan,inplace = True)
energy.drop(['Unnamed: 0','Unnamed: 1'], axis = 1, inplace = True)
energy['Country'] = energy.index
energy.rename(columns={'%':  '% Renewable', 'Petajoules': 'Energy Supply','Gigajoules': 'Energy Supply per Capita'},
                  inplace=True)
energy['Energy Supply'] = 100000*energy['Energy Supply']
energy['index'] = range(0,len(energy))
energy = energy.set_index('index')
energy['Country'] = energy['Country'].str.replace('\d+', '')
energy.head()
```
